```python
from opyrability import multimodel_rep, OI_eval
import numpy as np
```

```python
def shower_problem_2x2(u):
    y = np.zeros(2)
    y[0]=u[0]+u[1]
    if y[0]!=0:
        y[1]=(u[0]*60+u[1]*120)/(u[0]+u[1])
    else:
        y[1]=(60+120)/2

    return y
```

```python
#Definindo os limites AIS e DOS:
DOS_bounds =  np.array([[5, 15],
                        [90, 100]])

AIS_bounds =  np.array([[1, 7.5],
                        [1, 7.5]])

AIS_resolution =  [5, 5]

model =  shower_problem_2x2
```

```python
#AOS pela abordagem multimodelo
AOS_region  =  multimodel_rep(model, AIS_bounds, AIS_resolution)
```

```python
OI = OI_eval(AOS_region, DOS_bounds)
```
